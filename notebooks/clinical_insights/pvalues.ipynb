## Models

| Method | Significant | Not Significant | Total |
|---|---|---|---|
| t-test | 0 | 2 | 2 |
| chi-square | 5 | 73 | 78 |
| ANOVA | 1 | 1 | 2 |
| Kruskal-Wallis | 1 | 1 | 2 |
| OLS | 1 | 0 | 1 |
| llr_p | 2 | 8 | 10 |
| mlogit_p | 20 | 65 | 85 |
| Fisher's Exact | 0 | 30 | 30 |

```python
import matplotlib.pyplot as plt
import numpy as np

methods = ['Logistic Regression\n(LLR p-value)', 'Chi-Square', 'Multinomial Logistic\nRegression (LLR p-value)']
significant = [2, 5, 20]
not_significant = [8, 73, 65]

x = np.arange(len(methods))
width = 0.6

fig, ax = plt.subplots(figsize=(10, 5))

bars1 = ax.barh(x, significant, width, label='Significant (p<0.05)', color='steelblue')
bars2 = ax.barh(x, not_significant, width, left=significant, label='Not Significant', color='lightgray')

ax.set_ylabel('Method', fontsize=12)
ax.set_xlabel('Count', fontsize=12)
ax.set_title('Significant vs. Insignificant Features by Statistical Method', fontsize=13)
ax.set_yticks(x)
ax.set_yticklabels(methods, fontsize=10)
ax.legend(fontsize=10)

# Add count labels inside bars
for bar in bars1:
    w = bar.get_width()
    if w > 0:
        ax.text(bar.get_x() + w/2, bar.get_y() + bar.get_height()/2,
                str(int(w)), ha='center', va='center', fontsize=9, color='white', fontweight='bold')

for bar in bars2:
    w = bar.get_width()
    if w > 0:
        ax.text(bar.get_x() + w/2, bar.get_y() + bar.get_height()/2,
                str(int(w)), ha='center', va='center', fontsize=9, color='black')

plt.tight_layout()
plt.savefig('model_level_barplot_horizontal.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

methods = ['OLS', 'llr_p\n(Logit coefs)', 'mlogit_p\n(MNLogit coefs)']
significant = [1, 16, 25]
not_significant = [10, 72, 29]

x = np.arange(len(methods))
width = 0.4

fig, ax = plt.subplots(figsize=(8, 6))

bars1 = ax.bar(x, significant, width, label='Significant (p<0.05)', color='steelblue')
bars2 = ax.bar(x, not_significant, width, bottom=significant, label='Not Significant', color='lightgray')

ax.set_xlabel('Method', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Significant vs. Not Significant Coefficients by Method', fontsize=13)
ax.set_xticks(x)
ax.set_xticklabels(methods, fontsize=11)
ax.legend(fontsize=10)

# Add count labels inside bars
for bar in bars1:
    h = bar.get_height()
    if h > 0:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + h/2,
                str(int(h)), ha='center', va='center', fontsize=10, color='white', fontweight='bold')

for bar in bars2:
    h = bar.get_height()
    if h > 0:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + h/2,
                str(int(h)), ha='center', va='center', fontsize=10, color='black')

plt.tight_layout()
plt.savefig('coefficient_level_barplot.png', dpi=150, bbox_inches='tight')
plt.show()
```
