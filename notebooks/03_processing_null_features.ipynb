```python
import os
import pandas as pd
```

```python
# Paths
PROCESSED_FILE = os.path.join("..", "data", "processed", "cardio_onc_prostate_02processed.csv")
CLEANED_FILE = os.path.join("..", "data", "processed", "cardio_onc_prostate_03cleaned.csv")

# Load processed CSV
df = pd.read_csv(PROCESSED_FILE)
print(f"Original shape: {df.shape}")

# Threshold for missing values
threshold = 0.5

# Calculate fraction of missing values per column
missing_frac = df.isna().mean()

# Identify columns to remove
cols_to_remove = missing_frac[missing_frac > threshold].index.tolist()
print(f"Columns to remove ({len(cols_to_remove)}):")
print(cols_to_remove)


# Drop columns from dataframe
df_cleaned = df.drop(columns = cols_to_remove)
print(f"Shape after removing columns: {df_cleaned.shape}")

# Save cleaned dataframe
df_cleaned.to_csv(CLEANED_FILE, index = False)
print(f"Cleaned dataset saved to: {CLEANED_FILE}")
```
